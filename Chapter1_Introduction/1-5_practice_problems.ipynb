```python
def describe(x):
  print("타입 {}".format(x.type()))
  print("크기 {}".format(x.size()))
  print("값 {}".format(x))
```

```python
import torch

#1번 -> 2D tensor를 만들고 차원 0위치에 크기가 1인 차원을 추가

x = torch.randn(2, 3).view(1, 2, 3)
describe(x)

#2번 -> 이전에 만든 차원을 삭제.
x = x.view(2, 3)
describe(x)
```

```python
#3번. 범위가 [3, 7)이고 크기가 5 * 3인 랜덤 텐서

x = torch.rand(5, 3)
y = 3 + 4 * x
describe(y)
```

```python
# 4번 정규분포를 이용하여 텐서를 만들기.
x = torch.randn(5, 3)
describe(x)
```

```python
#5번. torch.Tensor([1, 1, 1, 0, 1])에서 0이 아닌 원소의 index를 추출

x = torch.Tensor([1, 1, 1, 0, 1])
describe(x)

print(torch.nonzero(x))
```

```python
#6번. 크기가 3, 1인 랜던한 텐서 4벌을 복사해 쌓기
x= torch.rand(3, 1)
x = torch.stack([x, x, x, x])
describe(x)
```

```python
#7번. 2차원 행렬 2개에 대해서 배치 행렬곱셈 수행

a = torch.rand(3, 4, 5)
b = torch.rand(3, 5, 4)

describe(a)
describe(b)
describe(torch.bmm(a, b))
```

```python
# 8번. 3차원 행렬과 2차원 행렬의 배치 행렬 곱셈을 계산하기

a = torch.rand(3, 4, 5)
b = torch.rand(5, 4)

batch_b = b.unsqueeze(0).expand(a.size(0), *b.size())
describe(batch_b)

c = torch.bmm(a, batch_b)
describe(c)
```
